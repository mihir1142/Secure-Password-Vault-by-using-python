# Project 1 Secure Password vault

```python
## Initial credentials list
credentials = [
    {"website": "speedup.com", "username": "chetana", "password": "Chetana123@"}
]

# Function to check password strength
def check_credentials(password):
    if len(password) < 8:
        return False, "Password must contain at least 8 characters"

    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif not char.isalnum():
            has_special = True

    if all([has_upper, has_lower, has_digit, has_special]):
        return True, "Password is strong"
    else:
        return False, "Password must contain uppercase, lowercase, digit, and special character"

# Function to add new credentials
def add_credentials():
    website = input("Enter website name: ")
    username = input("Enter user name: ")
    password = input("Enter password: ")

    valid, message = check_credentials(password)
    print(message)

    if valid:
        credentials.append({"website": website, "username": username, "password": password})
        print("Credentials saved successfully ✅")
    else:
        print("No credentials saved ❌")

# Function to view all credentials
def view_credentials():
    if credentials:
        print("Stored credentials:")
        for dictionary in credentials:
            print(dictionary)
    else:
        print("No credentials stored.")

# Function to search credentials by website
def search_credentials():
    site = input("Enter website to search: ")
    found = False
    for d in credentials:  # ✅ fixed (removed ())
        if d["website"].lower() == site.lower():
            print("Website found:", d)
            found = True
            break
    if not found:
        print("No website found")

# Menu-driven loop
while True:
    print("\nWelcome to Secure Password Vault!")
    print("1. Add new credential")
    print("2. View all credentials")
    print("3. Search by website")
    print("4. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        add_credentials()
    elif choice == "2":
        view_credentials()
    elif choice == "3":
        search_credentials()
    elif choice == "4":
        print("Exiting program!")
        break
    else:
        print("Invalid choice, please choose 1, 2, 3, or 4")
```
